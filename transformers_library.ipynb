```python
from datasets import load_dataset
squad = load_dataset("squad", split="train[:4999]")
```

```python
squad = squad.train_test_split(test_size=0.2)
```

```python
squad['test'][0]
```

```python
squad['train'][0]
```

```python
model_name = "distilbert/distilbert-base-uncased"
```

```python
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
def preprocess_function(samples):
  questions= [q.strip() for q in samples["question"]]
  inputs = tokenizer(
      questions,
      samples["context"],
      max_length=384,
      truncation="only_second",
      return_offsets_mapping=True,
      padding="max_length",
  )

  offset_mapping = inputs.pop("offset_mapping")
  answers = samples["answers"]
  start_positions = []
  end_positions=[]

  for i, offset in enumerate(offset_mapping):
    answer = answers[i]
    start_char = answer["answer_start"][0]
    end_char = start_char + len(answer["text"][0])
    sequence_ids = inputs.sequence_ids(i)

    idx = 0
    while sequence_ids[idx] !=1:
      idx +=1
    context_start = idx
    while sequence_ids[idx] == 1:
      idx +=1
    context_end = idx-1

    if offset[context_start][0] > end_char or offset[context_end][1] < start_char:
      start_positions.append(0)
      end_positions.append(0)
    else:
      idx = context_start
      while idx <= context_end and offset[idx][0] <= start_char:
        idx +=1
      start_positions.append(idx-1)

      idx = context_end
      while idx >= context_start and offset[idx][1] >= end_char:
        idx -=1
      end_positions.append(idx+1)

  inputs["start_positions"] = start_positions
  inputs["end_positions"] = end_positions
  return inputs
```

```python
# to see the differences
tokenized2 = squad.map(preprocess_function, batched=True)
```

```python
tokenized_squad = squad.map(preprocess_function, batched=True, remove_columns = squad["train"].column_names)
```

```python
from transformers import DefaultDataCollator

data_collator=DefaultDataCollator(return_tensors="tf")
```

```python
data_collator
```

```python
from transformers import create_optimizer

batch_size = 16
num_epochs = 2
total_train_steps = ( len(tokenized_squad["train"]) // batch_size) * num_epochs

optimizer, schedule = create_optimizer(
    init_lr=2e-5,
    num_warmup_steps=0,
    num_train_steps=total_train_steps,
)
```

```python
from transformers import TFAutoModelForQuestionAnswering

model= TFAutoModelForQuestionAnswering.from_pretrained(model_name)
```

```python
tf_train_set = model.prepare_tf_dataset(
    tokenized_squad["train"],
    shuffle = True,
    batch_size=16,
    collate_fn=data_collator,
)

tf_validation_set = model.prepare_tf_dataset(
    tokenized_squad["test"],
    shuffle=False,
    batch_size=16,
    collate_fn=data_collator,
)
```

```python
import tensorflow as tf
model.compile(optimizer=optimizer)
```

```python
from transformers.keras_callbacks import PushToHubCallback

callback = PushToHubCallback(
    output_dir="first_model_on_the_hub",
    tokenizer=tokenizer,
)
```

```python
model.fit(x=tf_train_set, validation_data =tf_validation_set,
          epochs=2,
          callbacks=[callback],
          verbose=2
          )
```
